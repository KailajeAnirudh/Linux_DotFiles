```python
import imageio
import matplotlib.pyplot as plt
import os
import numpy as np

plt.figure()
ax = plt.gca()
gif_path = "/home/anirudhkailaje/Documents/01_UPenn/02_MEAM5170/03_FinalProject/report/Biped Hop/ppt/media/leg_1.gif"
animation = imageio.get_reader(gif_path)
frames = []
for i, frame in enumerate(animation):
    if i==110:
        frames.append(frame)
    if i>40:
        continue
    if (i%15==0 and i>0):
        frames.append(frame)

import cv2
cropheight = 160
cropwidth = 160
h = cropheight
w = cropwidth
image = np.zeros((h, w*3, 3))
for i, f in enumerate(frames):
    f = f[320-(h//2):320+h//2,320-w//2:320+w//2,:3]
    image[:,i*(w):(i+1)*(w)] = f
cv2.imwrite("output.bmp", image)
```
