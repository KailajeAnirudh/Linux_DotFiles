```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
import pickle

with open('/home/anirudhkailaje/Documents/01_UPenn/01_ESE5460/02_Homework/dels.pickle', 'rb') as f:
    train_losses, del_ins, del_outs = pickle.load(f)

del_ins = np.log(del_ins)
del_outs = np.log(del_outs)

ns = np.logspace(1, 3, 20).astype(int)

# Create figure and axis objects
fig, ax1 = plt.subplots()

# Plot del_ins on the first y-axis
bp1 = ax1.boxplot(del_ins.T, positions=ns, patch_artist=True)
ax1.set_xscale('log')  # Set x-axis to log scale
ax1.set_xlabel('Labels (ns)')
ax1.set_ylabel('del_ins', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Create a second y-axis
ax2 = ax1.twinx()
bp2 = ax2.boxplot(del_outs.T, positions=ns, patch_artist=True)
ax2.set_ylabel('del_outs', color='tab:orange')
ax2.tick_params(axis='y', labelcolor='tab:orange')

# Customize the boxplots
colors = ['lightblue', 'lightgreen', 'lightyellow', 'lightpink', 'lightcoral']
for box in bp1['boxes']:
    box.set_facecolor(colors[0])
for box in bp2['boxes']:
    box.set_facecolor(colors[1])

# Add a legend
plt.legend([bp1["boxes"][0], bp2["boxes"][0]], ['del_ins', 'del_outs'], loc='upper left')

# Show the plot
plt.show()

```

```python
fig, ax1 = plt.subplots()
bp1 = ax1.boxplot(del_ins.T, labels=ns, patch_artist=True)
ax1.set_ylabel('$\delta_{in}(n)$')
ax1.set_xlabel('$n$')
ax1.tick_params(axis='y', labelcolor = 'lightblue')
ax2 = ax1.twinx()
bp2 = ax2.boxplot(del_outs.T, labels=ns, patch_artist = True)
ax2.set_ylabel('$\delta_{out}(n)$')
ax2.tick_params(axis='y', labelcolor = 'orange')

for box in bp1['boxes']:
    box.set_facecolor('lightblue')
    box.set_alpha(0.35)
for line in bp1['medians']:
    line.set_color('blue')
for box in bp2['boxes']:
    box.set_facecolor('orange')
    box.set_alpha(0.35)
for line in bp2['medians']:
    line.set_color('red')
for circle in bp2['fliers']:
    circle.set_color('red')
plt.show()
```
