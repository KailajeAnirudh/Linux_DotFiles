```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('2024A MEAM543 UAV Database  - Sheet1.csv', skiprows=1)
df.columns = [col.replace('/','').strip()
              .replace('\n', '_').replace(' ','_')
              .replace('[','').replace(']','')
               .replace('(','').replace(')','').replace("\"","") for col in df.columns.to_list()]
df
```

```python
import re
def get_numbers(x):
    num_float = re.findall(r'\d+\.\d+', str(x))
    dec_float = re.findall(r'\d+', str(x))
    comma_seperated_dec = re.findall(r'\d+\,\d+', str(x))
    comma_seperated_float = re.findall(r'\d+\,\d+\.\d+', str(x))
    if num_float:
        return np.array([float(num) for num in num_float]).max()
    elif dec_float:
        return np.array([float(num) for num in dec_float]).max()
    elif comma_seperated_dec:
        return np.array([float(num.replace(',','')) for num in comma_seperated_dec]).max()
    elif comma_seperated_float:
        return np.array([float(num.replace(',','')) for num in comma_seperated_float]).max()
    else:
        return np.nan
    
```

```python
df.Max_Payload_Weight_lb = df.Max_Payload_Weight_lb.apply(get_numbers)
df.Max_Flight_Distance_Range_miles = df.Max_Flight_Distance_Range_miles.apply(get_numbers)
df.Cruise_Altitude_ft = df.Cruise_Altitude_ft.apply(get_numbers)
Payload_mask = (df.Max_Payload_Weight_lb > (3.3*0.75)) & (df.Max_Payload_Weight_lb < (3.3*1.25))
Range_mask = (df.Max_Flight_Distance_Range_miles > (30*0.55)) & (df.Max_Flight_Distance_Range_miles < (30*1.25))
df[Payload_mask & Range_mask]
```
